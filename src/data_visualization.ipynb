```python
#implelmenting histogram 
data=pd.read_csv("D:\DownLoad\projects\Auto_learn\data\cleaned_from_outliers_df.csv")
df=pd.DataFrame(data)
numeric_columns=df.select_dtypes(include=['int','float']).hist(bins=10,figsize=(20,10))
plt.tight_layout()
plt.show()
```

```python
#Boxplot Visualization
data=pd.read_csv("D:\DownLoad\projects\Auto_learn\data\cleaned_from_outliers_df.csv")
df=pd.DataFrame(data)
numeric_col=df.select_dtypes(include=['int','float']).columns
plt.figure(figsize=(20, len(numeric_col) * 3))
for i, col in enumerate(numeric_col, 1):
    plt.subplot(len(numeric_col), 1, i)
    sns.boxplot(x=df[col], color='salmon')
    plt.title(f'Box Plot of {col}')
    plt.tight_layout()
plt.show()
```

```python
data=pd.read_csv("D:\DownLoad\projects\Auto_learn\data\cleaned_from_outliers_df.csv")
df=pd.DataFrame(data)
df=df.select_dtypes(include=['int','float'])
correlation_matrix = df.corr()


plt.figure(figsize=(18, 14))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=18)
plt.show()
```
